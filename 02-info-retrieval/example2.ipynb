```python
# Sample project
import nltk

sentence = """ At eight o'clock on Thursday morning 
Arthor didn't feel very good"""
```

```python
# tokenize some text
tokens = nltk.word_tokenize(sentence)
tokens
```

```python
# Tag text
tagged = nltk.pos_tag(tokens)
tagged[0:6]
```

```python
# Identify entities
entities = nltk.chunk.ne_chunk(tagged)
entities
```
